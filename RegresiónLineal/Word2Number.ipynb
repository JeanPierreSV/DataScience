```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
d = pd.read_csv("hiring.csv")
d
```

```python
d.experience = d.experience.fillna("zero")
d
```

```python
d.experience = d.experience.apply(w2n.word_to_num)
d
```

```python
import math
median_test_score = math.floor(d['test_score(out of 10)'].mean())
median_test_score
```

```python
d['test_score(out of 10)'] = d['test_score(out of 10)'].fillna(median_test_score)
d
```

```python
reg = linear_model.LinearRegression()
reg.fit(d[['experience','test_score(out of 10)','interview_score(out of 10)']],d['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Cuánto será el salario para una persona con 2 años de experiencia, 
# si obtuvo un puntaje en el test de 9 y en la entrevista de 6
reg.predict([[2,9,6]])
```

```python
x1 = 2922.26901502*2 + 2221.30909959*9 + 2147.48256637*6 + 14992.65144669314
print(x1)
```
