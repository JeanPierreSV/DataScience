```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

## **Pickle**

<h3 style='color:purple'>Save Model To a File Using Python Pickle</h3>

```python
# Guardamos el archivo en formato pickle con el nombre de 'model_pickle'
import pickle
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

<h4 style='color:purple'>Load Saved Model</h4>

```python
# Abrimos el archivo 'model_pickle' y le damos el nombre de mp
with open('model_pickle','rb') as file:
    mp = pickle.load(file)
```

```python
mp.coef_
```

```python
mp.intercept_
```

```python
mp.predict([[5000]])
```

## **joblib**

<h3 style='color:purple'>Save Trained Model Using joblib</h3>

```python
# Guardamos el archivo en formato joblib con el nombre de 'model_joblib'
from sklearn.externals import joblib
joblib.dump(model, 'model_joblib')
```

<h4 style='color:purple'>Load Saved Model</h4>

```python
# Abrimos el archivo 'model_joblib' y le damos el nombre de mj
mj = joblib.load('model_joblib')
```

```python
mj.coef_
```

```python
mj.intercept_
```

```python
mj.predict([[5000]])
```
